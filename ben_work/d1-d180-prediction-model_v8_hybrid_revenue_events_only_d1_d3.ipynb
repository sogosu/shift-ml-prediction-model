## Step 27: ### Data Processing Step


**Data Cleaning**: This removes unnecessary or problematic data. Cleaning might involve dropping duplicate records, removing outliers, or eliminating features that don't contribute to predictions. Clean data leads to more accurate models.

```python
# Custom binning function with more granular bins for hours_active
def custom_binning(series, q=4, labels=["very_short", "short", "medium", "long"], zero_label="zero"):
    series = series.fillna(0)
    if series.name == "hours_active":
        series = series.clip(lower=0)
    zero_mask = series == 0
    non_zero_series = series[~zero_mask]
    if len(non_zero_series) == 0:
        return pd.Series(zero_label, index=series.index, dtype="object")
    try:
        bins = pd.qcut(non_zero_series, q=q, labels=labels, duplicates="drop")
        result = pd.Series(zero_label, index=series.index, dtype="object")
        result.loc[~zero_mask] = bins.reindex(non_zero_series.index).astype("object")
        return result
    except ValueError as e:
        print(f"pd.qcut failed for {series.name}: {e}")
        quantiles = non_zero_series.quantile([0, 0.25, 0.5, 0.75, 1.0])
        unique_quantiles = sorted(set(quantiles))
        bin_edges = [0] + unique_quantiles
        num_bins = len(bin_edges) - 1
        if num_bins < len(labels) + 1:
            adjusted_labels = [zero_label] + labels[:num_bins - 1]
        else:
            adjusted_labels = [zero_label] + labels
        if len(bin_edges) < 2:
            print(f"Only one unique value in non-zero {series.name}, assigning all non-zero to first label")
            result = pd.Series(zero_label, index=series.index, dtype="object")
            result[~zero_mask] = labels[0]
            return result
        try:
            return pd.cut(series, bins=bin_edges, labels=adjusted_labels, include_lowest=True)
        except ValueError as e:
            print(f"pd.cut also failed for {series.name}: {e}")
            if len(set(non_zero_series)) == 1:
                result = pd.Series(zero_label, index=series.index, dtype="object")
                result[~zero_mask] = labels[0]
                return result
            else:
                min_val, max_val = non_zero_series.min(), non_zero_series.max()
                bin_edges = [0, min_val, min_val + (max_val - min_val) / 3, min_val + 2 * (max_val - min_val) / 3, max_val + 1]
                bin_edges = sorted(set(bin_edges))
                num_bins = len(bin_edges) - 1
                adjusted_labels = [zero_label] + labels[:num_bins - 1]
                return pd.cut(series, bins=bin_edges, labels=adjusted_labels, include_lowest=True)
```

**Processing Logic**: This code performs calculations or transformations on the data. Each transformation prepares the data for machine learning by extracting meaningful patterns or formatting data appropriately for algorithms.

```python
# Custom objective function for XGBoost to optimize Within ±10% (reverted to original)
def custom_within_10pct_objective(y_true, y_pred):
    y_true_orig = np.expm1(y_true)
    y_pred_orig = np.expm1(y_pred)
    within_10 = np.where(y_true_orig > 1,
                        np.abs((y_pred_orig - y_true_orig) / (y_true_orig + 1e-10)) <= 0.1,
                        np.abs(y_pred_orig - y_true_orig) <= 0.1)
    residual = y_pred - y_true
    grad = np.where(within_10, 0, 2 * residual)  # Zero gradient within 10%
    hess = np.where(within_10, 0, 2)  # Zero Hessian within 10%
    return grad, hess
```

```python
# Process day1_events: Extract individual events
all_events = []
for events in preferable_users['day1_events']:
    # Handle string representation of lists
    if isinstance(events, str):
        # Remove brackets and split by comma
        if events == '[]':
            continue  # Skip empty lists
        # Replace brackets and split
        events_clean = events.strip('[]').split(', ')
        all_events.extend([event.strip() for event in events_clean if event.strip()])
    elif isinstance(events, list):
        all_events.extend([str(event) for event in events if event])
```
